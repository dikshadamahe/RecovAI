```python
import pandas as pd

df = pd.read_csv("ML_Dataset_Copper_TARGET85.csv", header=1)

print(df.columns.tolist())
```

```python
df["Recovery (%)"].head()
```

Average Copper Recovery

```python
avg_recovery = df['Recovery (%)'].mean()
print("Average Recovery:", avg_recovery)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.histplot(df["Recovery (%)"], kde=True)
plt.title("Distribution of Copper Recovery (%)")
plt.show()
```

correlation with recovery


```python
numeric_df = df.select_dtypes(include="number")

corr_with_recovery = numeric_df.corr()["Recovery (%)"].sort_values(ascending=False)
corr_with_recovery
```

```python
plt.figure(figsize=(8, 10))
sns.heatmap(corr_with_recovery.to_frame(), annot=True, cmap="coolwarm")
plt.title("Correlation of Features with Recovery (%)")
plt.show()
```

```python
plt.figure(figsize=(18, 12))
sns.heatmap(numeric_df.corr(), cmap="coolwarm", center=0)
plt.title("Correlation Heatmap")
plt.show()
```

recovery vs important features

```python
important_features = [
    "Head Grade (%Cu)",
    "Tails Grade (%Cu)",
    "Feed Rate (MT/h)",
    "Flotation pH",
    "SIPX Dose (g/t)",
    "Frother Dose (g/t)",
    "Lime Bags",
    "Depressant Dose (g/t)",
    "Roll7_Recovery (%)",
    "Prev_Recovery (%)"
]

for col in important_features:
    plt.figure(figsize=(6, 4))
    sns.scatterplot(data=df, x=col, y="Recovery (%)")
    plt.title(f"Recovery (%) vs {col}")
    plt.show()
```

outlier detection


```python
numeric_cols = numeric_df.columns

for col in numeric_cols:
    plt.figure(figsize=(6, 3))
    sns.boxplot(x=df[col])
    plt.title(f"Boxplot of {col}")
    plt.show()
```

recovery by shift

```python
sns.boxplot(data=df, x="Shift", y="Recovery (%)")
plt.title("Recovery by Shift")
plt.show()
```

time and trend analyisis

```python
df["Date"] = pd.to_datetime(df["Date"], dayfirst=True, errors="coerce")

plt.figure(figsize=(14, 5))
sns.lineplot(data=df, x="Date", y="Recovery (%)")
plt.title("Recovery Trend Over Time")
plt.xticks(rotation=45)
plt.show()
```

reagent analysis

```python
reagent_cols = [
    "SIPX Dose (g/t)",
    "Frother Dose (g/t)",
    "Lime Bags",
    "T Reagent (cc)",
    "Pine Oil (cc)",
    "Depressant Dose (g/t)"
]

df[reagent_cols].describe()
```

```python
for col in reagent_cols:
    plt.figure(figsize=(6, 4))
    sns.scatterplot(data=df, x=col, y="Recovery (%)")
    plt.title(f"Recovery vs {col}")
    plt.show()
```
